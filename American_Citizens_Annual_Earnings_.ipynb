#**Dataset Information**

**Source:** https://www.gigasheet.com/sample-data/american-citizens-annual-incomecsv

**Year / Timeline:** Data collected during 2024

**Dataset Column Description:**

* **Workclass** - Employment type (e.g.,State-Government, Local-Government, Private, Self-employed income, Self-employed not income)
* **Age** - Age of the individual members.
* **Final weight** - Census weighting factor represents.
* **Education** - Highest level of education attained (e.g., Secondary school, High school, Bachelors,Masters).
* **Education.num** - Numeric representation of education level(e.g.,10 = Secondary school,12 = High school, 13= Bachelors).
* **Marital.status** - Marital status(e.g., Married,Never married, Divorced, Separated, Widowed).
* **Occupation** - Job role, industry sector (e.g.,Prof-speciality,Craft-repair,Adm-clerical,Sales,Tech-support,Other-service).
* **Relationship** - Family relationship status(e.g.,Wife,Husband,Own-child,Unmarried,Not-in-family.Other relative).
* **Race** - Race of the individual.
* **Sex** - Gender of the members(Male/Female).
* **Captial.gain** - Income from captial gains.
* **Captial.loss** - Losses from captial.
* **Hours.per.week** - Average number of working hours per week.
* **Native.country** - Country of origin.
* **Income** - Income category(e.g.,>50k or <=50k per year)

```python
df['Income.numeric'] = df['Income'].map({'>50K': 1, '<=50K': 0})
```

```python
print(df['Occupation'].unique())
```

```python
df['Native.Country'] = df['Native.Country'].replace('?', np.nan)
df['Native.Country'] = df['Native.Country'].fillna('Unknown')
```

```python
print(df['Native.Country'].unique())
```

```python
df['Workclass'] = df['Workclass'].replace('?', np.nan)
df['Workclass']=df['Workclass'].fillna(df['Workclass'].mode()[0])
```

```python
print(df['Workclass'].unique())
```

```python
bins = [0, 18, 30, 40, 60, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr

df = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]
```

```python
df.loc[:, 'Age_log'] = np.log1p(df['Age'])
```

```python
df.loc[:, 'Work_hours_group'] = pd.cut(df['Hours.Per.Week'],bins=[0, 20, 40, 60, 100],labels=['Part-time', 'Standard', 'Overtime', 'Extreme'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df with columns 'Hours.Per.Week' and 'Sex'
plt.figure(figsize=(8,6))
sns.boxplot(x='Sex', y='Hours.Per.Week', data=df, palette='Set2')

plt.title('Distribution of Hours Per Week by Sex')
plt.xlabel('Sex')
plt.ylabel('Hours Per Week')
plt.show()
```

**Interpretation:** It shows how weekly working hours differ between males and females.



**Business Insights:** Males generally work longer hours on average, while females display slightly lower median hours with overlapping variability.


```python

# Assuming your DataFrame is named df and has columns 'Education', 'Hours.Per.Week', and 'Sex'
plt.figure(figsize=(12,6))
sns.stripplot(x='Education', y='Hours.Per.Week', hue='Sex',
              data=df, dodge=True, jitter=True, alpha=0.7, palette='Set2')

plt.title('Scatter Plot of Education vs Hours Per Week by Sex')
plt.xlabel('Education')
plt.ylabel('Hours Per Week')
plt.xticks(rotation=45)  # Rotate labels for readability
plt.legend(title='Sex', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Interpretation:** It shows how weekly working hours vary across education levels for males and females.


**Business Insights:** Higher education levels tend to align with longer working hours, with noticeable differences between sexes across categories


```python

# Assuming your DataFrame is named df and has columns 'Age', 'Work_hours_group', and 'Relationship'
plt.figure(figsize=(12,6))
sns.barplot(x='Work_hours_group', y='Age', hue='Relationship',
            data=df, ci=None, palette='Set2')

plt.title('Comparison of Age across Work Hours Group, segmented by Relationship')
plt.xlabel('Work Hours Group')
plt.ylabel('Average Age')
plt.legend(title='Relationship', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

**Interpretation:** It shows how average age differs across work hour groups segmented by relationship status.


**Business Insights:** It reveals demographic patterns where relationship roles influence age distribution across varying work commitments.



```python
cols = ['Age', 'Education.Num', 'Hours.Per.Week']  # exclude Income since it's categorical
corr_matrix = df[cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Heatmap: Age, Education.Num, Hours.Per.Week')
plt.show()
```

**Interpretation:** It shows correlations among Age, Education.Num, and Hours.Per.Week.





**Business Insights:** These relationships reveal how education and age influence working hours, guiding workforce planning and productivity strategies.

```python
# Select variables
corr_vars = ['Age','Education.Num','Hours.Per.Week','Capital.Gain','Capital.Loss','Income.numeric']

# Compute correlation matrix
corr_matrix = df[corr_vars].corr()

# Plot using Matplotlib only
plt.figure(figsize=(8,6))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()

# Add annotations
for i in range(len(corr_matrix)):
    for j in range(len(corr_matrix)):
        plt.text(j, i, round(corr_matrix.iloc[i, j], 2),
                 ha='center', va='center', color='black')

# Configure ticks and labels
plt.xticks(range(len(corr_vars)), corr_vars, rotation=45)
plt.yticks(range(len(corr_vars)), corr_vars)
plt.title("Correlation Heatmap: Income Drivers")
plt.tight_layout()
plt.show()
```

**Interpretation:** It shows correlations among age, education, work hours, capital gains/losses, and income.


**Business Insights:** These relationships highlight key drivers of income, guiding strategies in workforce planning and financial analysis
